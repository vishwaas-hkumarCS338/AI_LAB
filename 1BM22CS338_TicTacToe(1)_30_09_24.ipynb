```python
#LAB01 - 30/09/2024
#Q1(A)
#TIC_TAC_TOE GAME USING MACHINE THAT IS INTELLIGENT USING MINIMAX FN

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board):
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]

    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]
    return None

def is_full(board):
    return all(cell != " " for row in board for cell in row)

def minimax(board, depth, is_maximizing):
    winner = check_winner(board)
    if winner == "O":  #Computer
        return 10 - depth
    elif winner == "X":  #Player
        return depth - 10
    elif is_full(board):
        return 0  #Draw

    if is_maximizing:
        best_score = -float('inf')
        for r in range(3):
            for c in range(3):
                if board[r][c] == " ":
                    board[r][c] = "O"  #Computer move
                    score = minimax(board, depth + 1, False)
                    board[r][c] = " "
                    best_score = max(best_score, score)
        return best_score
    else:
        best_score = float('inf')
        for r in range(3):
            for c in range(3):
                if board[r][c] == " ":
                    board[r][c] = "X"  #Player move
                    score = minimax(board, depth + 1, True)
                    board[r][c] = " "
                    best_score = min(best_score, score)
        return best_score

def computer_move(board):
    best_score = -float('inf')
    best_move = (-1, -1)

    for r in range(3):
        for c in range(3):
            if board[r][c] == " ":
                board[r][c] = "O"  #Computer move
                score = minimax(board, 0, False)
                board[r][c] = " "
                if score > best_score:
                    best_score = score
                    best_move = (r, c)

    return best_move

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]

    current_player = "X"  #Player X is the human

    while True:
        print_board(board)

        if current_player == "X":
            row = int(input("Player X, enter your row (0-2): "))
            col = int(input("Player X, enter your column (0-2): "))
        else:
            row, col = computer_move(board)
            print("Computer (O) chooses row " + str(row) + ", column " + str(col))

        if board[row][col] == " ":
            board[row][col] = current_player
        else:
            print("Cell already taken! Try again.")
            continue

        winner = check_winner(board)
        if winner:
            print_board(board)
            print("Player " + winner + " wins!")
            break

        if is_full(board):
            print_board(board)
            print("It's a draw!")
            break

        current_player = "O" if current_player == "X" else "X"

tic_tac_toe()2
```
