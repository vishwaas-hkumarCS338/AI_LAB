```python
#LAB01 - 30/09/2024
#Q1(A)
#TIC_TAC_TOE GAME USING MACHINE THAT IS INTELLIGENT USING MINIMAX FN

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board):
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]

    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]
    return None

def is_full(board):
    return all(cell != " " for row in board for cell in row)

def minimax(board, depth, is_maximizing):
    winner = check_winner(board)
    if winner == "O":  #Computer
        return 10 - depth
    elif winner == "X":  #Player
        return depth - 10
    elif is_full(board):
        return 0  #Draw

    if is_maximizing:
        best_score = -float('inf')
        for r in range(3):
            for c in range(3):
                if board[r][c] == " ":
                    board[r][c] = "O"  #Computer move
                    score = minimax(board, depth + 1, False)
                    board[r][c] = " "
                    best_score = max(best_score, score)
        return best_score
    else:
        best_score = float('inf')
        for r in range(3):
            for c in range(3):
                if board[r][c] == " ":
                    board[r][c] = "X"  #Player move
                    score = minimax(board, depth + 1, True)
                    board[r][c] = " "
                    best_score = min(best_score, score)
        return best_score

def computer_move(board):
    best_score = -float('inf')
    best_move = (-1, -1)

    for r in range(3):
        for c in range(3):
            if board[r][c] == " ":
                board[r][c] = "O"  #Computer move
                score = minimax(board, 0, False)
                board[r][c] = " "
                if score > best_score:
                    best_score = score
                    best_move = (r, c)

    return best_move

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]

    current_player = "X"  #Player X is the human

    while True:
        print_board(board)

        if current_player == "X":
            row = int(input("Player X, enter your row (0-2): "))
            col = int(input("Player X, enter your column (0-2): "))
        else:
            row, col = computer_move(board)
            print("Computer (O) chooses row " + str(row) + ", column " + str(col))

        if board[row][col] == " ":
            board[row][col] = current_player
        else:
            print("Cell already taken! Try again.")
            continue

        winner = check_winner(board)
        if winner:
            print_board(board)
            print("Player " + winner + " wins!")
            break

        if is_full(board):
            print_board(board)
            print("It's a draw!")
            break

        current_player = "O" if current_player == "X" else "X"

tic_tac_toe()
```

```python
#LAB01 - 30/09/2024
#Q1(B)
#TIC_TAC_TOE GAME USING MACHINE THAT ACTIVELY ATTEMPTS TO BLOCK USER FROM WINNING

import random
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board):
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return board[0][i]

    if board[0][0] == board[1][1] == board[2][2] != " ":
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return board[0][2]

    return None

def block_win(board):
    # Check rows and columns for blocking
    for i in range(3):
        # Check rows
        if board[i].count("X") == 2 and board[i].count(" ") == 1:
            return (i, board[i].index(" "))
        # Check columns
        column = [board[j][i] for j in range(3)]
        if column.count("X") == 2 and column.count(" ") == 1:
            return (column.index(" "), i)

    # Check diagonals
    if board[0][0] == board[1][1] == "X" and board[2][2] == " ":
        return (2, 2)
    if board[0][2] == board[1][1] == "X" and board[2][0] == " ":
        return (2, 0)

    return None

def is_full(board):
    return all(cell != " " for row in board for cell in row)

def computer_move(board):
    # First try to block the player from winning
    block = block_win(board)
    if block:
        return block  # Return the blocking move

    # If no blocking move, pick a random empty cell
    empty_cells = [(r, c) for r in range(3) for c in range(3) if board[r][c] == " "]
    return random.choice(empty_cells)

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]

    current_player = "X"  # Player X is the human

    while True:
        print_board(board)

        if current_player == "X":
            row = int(input("Player X, enter your row (0-2): "))
            col = int(input("Player X, enter your column (0-2): "))
        else:
            row, col = computer_move(board)
            print("Computer (O) chooses row " + str(row) + ", column " + str(col))

        if board[row][col] == " ":
            board[row][col] = current_player
        else:
            print("Cell already taken! Try again.")
            continue

        winner = check_winner(board)
        if winner:
            print_board(board)
            print("Player " + winner + " wins!")
            break

        if is_full(board):
            print_board(board)
            print("It's a draw!")
            break

        current_player = "O" if current_player == "X" else "X"

tic_tac_toe()
```

```python
#LAB01 - 30/09/2024
#Q2
#VACUUM CLEANER PROBLEM

import random

# Define the environment as a dictionary
# Each key is a location, and the value is whether the location is 'Dirty' or 'Clean'
environment = {
    "A": random.choice(["Dirty", "Clean"]),
    "B": random.choice(["Dirty", "Clean"])
}

# Initial location of the vacuum cleaner
vacuum_location = random.choice(["A", "B"])

# Function to display the environment and agent's location
def display_environment():
    print("Location A is", environment['A'])
    print("Location B is", environment['B'])
    print("Vacuum is at location", vacuum_location)
    print()

# Define the agent function
def vacuum_agent():
    global vacuum_location

    # Loop until both locations are clean
    while environment["A"] == "Dirty" or environment["B"] == "Dirty":
        display_environment()

        # Check the current location
        if environment[vacuum_location] == "Dirty":
            # Clean the current location
            print("Cleaning location", vacuum_location)
            print()
            environment[vacuum_location] = "Clean"
        else:
            # Move to the other location
            print("Location", vacuum_location, "is clean. Moving to the other location...")
            print()
            vacuum_location = "A" if vacuum_location == "B" else "B"

    print("Both locations are clean. Task completed!")

# Run the vacuum cleaner agent
vacuum_agent()
```
